```
def mobility_graph(mobility_parameter):
  folder = 'drive/MyDrive/'
  mobility_data = pd.read_csv(folder+'2020_US_Region_Mobility_Report.csv')
  #mobility_data = mobility_data.sort_index(inplace=False, ascending=True)

  scaler = MinMaxScaler(feature_range = [-1, 1])
  mobility_data[[mobility_parameter['attribute']]] = scaler\
    .fit_transform(mobility_data[[mobility_parameter['attribute']]].values.reshape(-1,1))
  
  
  plt.plot(mobility_data[[mobility_parameter['attribute']]], mobility_parameter['color'], label=mobility_parameter['label'])

  plt.xlabel('Days(0-278),(last Nov 23rd)')
 
  
  
  plt.ylabel('Mobility data of '+mobility_parameter['label'])
  plt.title('Plot for '+mobility_parameter['label']+' in US')
  

  plt.figlegend(loc='upper left')
```

```
mobility_parameter = {
    'attribute' : 'retail_and_recreation_percent_change_from_baseline',
    'color' : 'green',
    'label' : 'Retail mobility'
}

mobility_graph(mobility_parameter)
```

```
mobility_parameter = {
    'attribute' : 'grocery_and_pharmacy_percent_change_from_baseline',
    'color' : 'red',
    'label' : 'Grocery and Pharmacies mobility'
}

mobility_graph(mobility_parameter)
```
