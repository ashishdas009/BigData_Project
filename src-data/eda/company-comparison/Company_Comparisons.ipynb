```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```
def compare_companies(company_1_data, company_2_data, company_3_data):
  folder = "drive/MyDrive/"
  company_1 = pd.read_csv(folder+company_1_data['cmp_name']+'.csv')
  company_2 = pd.read_csv(folder+company_2_data['cmp_name']+'.csv')
  company_3 = pd.read_csv(folder+company_3_data['cmp_name']+'.csv')

  company_1= company_1.sort_index(inplace=False, ascending=True)
  scaler = MinMaxScaler(feature_range = [-1, 1])
  company_1[['ClosePrice']] = scaler\
    .fit_transform(company_1[['ClosePrice']].values.reshape(-1,1))

  company_2= company_2.sort_index(inplace=False, ascending=True)
  scaler = MinMaxScaler(feature_range = [-1, 1])
  company_2[['ClosePrice']] = scaler\
    .fit_transform(company_2[['ClosePrice']].values.reshape(-1,1)) 
  
  company_3= company_3.sort_index(inplace=False, ascending=True)
  scaler = MinMaxScaler(feature_range = [-1, 1])
  company_3[['ClosePrice']] = scaler\
    .fit_transform(company_3[['ClosePrice']].values.reshape(-1,1))
  

  plt.rc('font', size=12)
  fig, ax = plt.subplots(figsize=(10, 6))

  plt.rcdefaults()
  ax.plot(company_1[['ClosePrice']], color='tab:'+company_1_data['color'], label=company_1_data['label'])
  ax.plot(company_2[['ClosePrice']],  color='tab:'+company_2_data['color'], label=company_2_data['label'])
  ax.plot(company_3[['ClosePrice']],  color='tab:'+company_3_data['color'], label=company_3_data['label'])

  ax.set_xlabel('Days(0-278),(last Nov 23rd)')
 
  
  
  ax.set_ylabel('Closing price (in dollars $)')
  ax.set_title('Stock closing price of '+company_1_data['label']+' vs '+company_2_data['label'] +' vs '+company_3_data['label'])
  ax.grid()

  ax.legend(loc='upper left')
```

Tech Industry

```
cmp_1 = { 
   'color': 'green',
   'label': 'Amazon',
   'cmp_name': 'AMZN' 
}
cmp_2 = { 
   'color': 'blue',
   'label': 'Google',
   'cmp_name': 'GOOG' 
}
cmp_3 = {
    'color': 'red',
    'label': 'Apple',
    'cmp_name': 'AAPL'
}
compare_companies(cmp_1, cmp_2, cmp_3)
```
